```python
import os

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
base = "./baza/"
class_path = "utwardzona_jpg/"
```

```python
#changing all directory from png to jpg
file_list = os.listdir(base+class_path)
file_list
```

```python
path_good = "Utwardzona/"
```

```python
os.makedirs(base+path_good)
```

```python
i = 0
for file in file_list:
    newfilename = str(i)+".png"
    i=i+1
    image = Image.open(base+class_path+file)
    #image = image.convert('L')
    new_image = image.resize((220,220))
    new_image.save(base+path_good+newfilename)
```
